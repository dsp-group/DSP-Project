# Calculate Responsive Points

```python
approved_hire['Responsive_Point'] = 0.0
count_value_null('Responsive_Point', approved_hire)
```

Create function to calculate responsive points for confirmed hires from actions

```python
def update_responsive_points(feature, point):
    for index, rowData in approved_hire.iterrows():
        if rowData[feature] == 1.0:
            approved_hire.at[index,'Responsive_Point'] = rowData['Responsive_Point'] + point
    print(approved_hire['Responsive_Point'].value_counts().sort_index())
```

```python
# True / False (indicates new resume created / revised)
update_responsive_points('Finalized_HHUSA_revised_resume_on_file__c', 1.5)    
```

```python
# True / False (indicates client has created a profile on HHUSA job board)
update_responsive_points('On_Job_Board__c', 1)           
```

```python
# Label cases as Easy/ Medium/ Hard
approved_hire['Case_Type'] = "Medium"
for index, rowData in approved_hire.iterrows():
    if (rowData['Responsive_Type'] == "Very Responsive") & (rowData['Highest_Level_of_Education_Completed__c'] == "4 Year Degree (BA, BS, etc.)" ):
        approved_hire.at[index,'Case_Type'] = "Easy"
    if (rowData['Responsive_Type'] == "Responsive") & (rowData['Highest_Level_of_Education_Completed__c'] == "Post-Graduate Degree (MA, MS, JD, etc.)" ):
        approved_hire.at[index,'Case_Type'] = "Easy"
    if (rowData['Responsive_Type'] == "Responsive") & (rowData['Highest_Level_of_Education_Completed__c'] == "4 Year Degree (BA, BS, etc.)" ) & (rowData['Security_Clearance_Description__c'] == "Yes" ):
        approved_hire.at[index,'Case_Type'] = "Easy"
    if (rowData['Responsive_Type'] == "Responsive") & (rowData['Highest_Level_of_Education_Completed__c'] == "High School/GED"):
        approved_hire.at[index,'Less Responsive'] = "Hard"
    if (rowData['Responsive_Type'] == "Responsive") & (rowData['Security_Clearance_Description__c'] == "No" ):
        approved_hire.at[index,'Case_Type'] = "Hard"
    if (rowData['Responsive_Type'] == "Less Responsive"):
        approved_hire.at[index,'Case_Type'] = "Hard"
print(approved_hire['Case_Type'].value_counts())
```

```python
# The total number of cases assgined to each owner
total = pd.crosstab(client['Case_Owner'], columns=['OwnerId'])
total.rename(columns={'OwnerId':'Total_Assgined'},inplace=True)
total.head(10)
```

```python
#add in the total number of cases assgined to each owner to the dataframe
successful["Total_Successful"] = successful["Easy"]+ successful["Hard"]+ successful["Medium"]
successful = pd.merge(successful, 
                      total, 
                      on= 'Case_Owner', 
                      how= 'left')
```

```python
# Only compare the experienced employees who completed more than 100 cases.
successful = successful[successful['Total_Successful']>100]
successful.head(10)
```

```python
# Calculate the percentage of Easy/ Medium/ Hard completed out of the total SUCCESSFUL cases for each owner
successful["Easy_P"] = round(successful["Easy"]/successful["Total_Successful"],2)
successful["Medium_P"] = round(successful["Medium"]/successful["Total_Successful"],2)
successful["Hard_P"] = round(successful["Hard"]/successful["Total_Successful"],2)
top_successful = successful[['Hard_P','Medium_P', 'Easy_P']].sort_values(by=['Hard_P'], ascending=False)
top_successful.head(10)
```

```python
# Calculate the percentage of Easy/ Medium/ Hard completed out of the total ASSGINED cases for each owner
successful["Easy_P"] = round(successful["Easy"]/successful["Total_Assgined"],2)
successful["Medium_P"] = round(successful["Medium"]/successful["Total_Assgined"],2)
successful["Hard_P"] = round(successful["Hard"]/successful["Total_Assgined"],2)
result = successful[['Hard_P','Medium_P', 'Easy_P']].sort_values(by=['Hard_P'], ascending=False)
result['Incompleted_P'] = 1 - result['Easy_P'] - result['Medium_P']- result['Hard_P']
result = result.head(10).sort_values(by=['Hard_P'], ascending=False)
result
```
